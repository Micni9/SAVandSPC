```python
import numpy as np
import matplotlib.pyplot as plt
from init2 import *
from Methods import SAV2,N2_SAV1b,ESAVII,SC
```

```python
Times = np.array([64,128,256])*(e_time-s_time)
dir = "Report"
```

```python
now = 1
for times in Times:
    dt_inv = (int)(times/(e_time-s_time))
    dt = 1/dt_inv
    Nx = times
    x = np.linspace(left,right,Nx,endpoint = False)
    phi_init = initial(x)
    C0 = 0
    
    (mE_SAV2,E_SAV2,E1,r,T) = SAV2(phi_init,Nx,times,s_time,e_time,left,right,C0)
    
    path1 = f"{dir}/2dSAV2_E{now}.png"
    path5 = f"{dir}/2dSAV2_dif{now}.png"
    
    plt.title(f'Energy at dt = 1/{dt_inv}')
    plt.plot(T,E_SAV2[:-1],'-x')
    plt.xlabel("t")
    plt.ylabel("E")
    plt.savefig(path1)
    plt.clf()

    plt.title(f'r^2-E1 at dt = 1/{dt_inv}')
    plt.plot(T,(r*r-E1)[:-1],'-x')
    plt.xlabel("t")
    plt.ylabel("r^2-E1")
    plt.savefig(path5)
    plt.clf()
    now += 1
```

```python
now = 1
for times in Times:
    dt_inv = (int)(times/(e_time-s_time))
    dt = 1/dt_inv
    Nx = times
    x = np.linspace(left,right,Nx,endpoint = False)
    phi_init = initial(x)
    
    C0 = 0
    
    (mE_NSAV1b,E_NSAV1b,T) = N2_SAV1b(phi_init,Nx,times,s_time,e_time,left,right,C0)

    path1 = f"{dir}/NSAVb2_E{now}.png"
    path3 = f"{dir}/NSAVb2_dif{now}.png"
    
    plt.plot(T,E_NSAV1b[:-1],'-x')
    plt.title(f'Energy at dt = 1/{dt_inv}')
    plt.xlabel("t")
    plt.ylabel("E")
    plt.savefig(path1)
    plt.clf()

    E = np.array(E_NSAV1b[:-1])
    q = np.array(mE_NSAV1b[:-1])
    plt.plot(T,q-E,'-x')
    plt.title(f'q-E at dt = 1/{dt_inv}')
    plt.xlabel("t")
    plt.ylabel("E")
    plt.savefig(path3)
    plt.clf()
    now += 1
```
